## Q1.Given the following requirement write a python program using OOPs
 

*Include  salary as 4rth attribute in Employee class
*Note: contructor method init is common in all classes  and is used to  initialize the datamembers


### 1

Employee class methods:
      Calculate salary: displays the message “ This method will be overridden”
      Display : displays first name, lastname and social security number  


```python
class Employee:
    def __init__(self,fName,lName,socialSecurityNumber):
        if isinstance(fName,str) and isinstance(lName,str) and isinstance(socialSecurityNumber,str):
            self.fName=fName
            self.lName=lName
            self.socialSecurityNumber=socialSecurityNumber
            self.salary=''
        else:
            raise TypeError('Employee class expects str value')
    def CalculateSalary(self,salary):
        return 'This method will be overridden'
    def Display(self):
        return f'{self.fName} {self.lName} {self.socialSecurityNumber}'
```

### 2

CommissionEmployee   methods:
     Calculate salary: calculates salary as product of grossSales and commissionRate
     Display : calls the display method in parent class using super keyword and also
                     displays  salary for CommissionEmployee       


```python
class ComissionEmployee(Employee):
    def __init__(self,fName,lName,socialSecurityNumber,grossRate,commissionRate):
        super().__init__(fName,lName,socialSecurityNumber)
        self.grossRate=grossRate
        self.commissionRate=commissionRate
    def CalculateSalary(self):
        self.salary = self.grossRate*self.commissionRate
        return self.salary
    def Display(self):
        return f'{super().Display()} {self.CalculateSalary()}'
```

### 3

HourlyEmployee   methods:
     Calculate salary: calculates salary as product of wage and hours
     Display : calls the display method in parent class using super keyword and also
                     displays  salary for HourlyEmployee


```python
class HourlyEmployed(Employee):
    def __init__(self,fName,lName,socialSecurityNumber,wage,hour):
        super().__init__(fName,lName,socialSecurityNumber)
        self.wage=wage
        self.hour=hour
    def CalculateSalary(self):
        self.salary = self.wage*self.hour
        return self.salary
    def Display(self):
        return f'{super().Display()} {self.CalculateSalary()}'
```

### 4


   SalariedEmployee   methods:
             Calculate salary: calculates salary same as  weekly salary 
             Display : calls the display method in parent class using super keyword and also
             displays  salary for SalariedEmployee


```python
class SalariedEmployee(Employee):
    def __init__(self,fName,lName,socialSecurityNumber,weeklySalary):
        super().__init__(fName,lName,socialSecurityNumber)
        self.weeklySalary=weeklySalary
    def CalculateSalary(self):
        salary=self.weeklySalary
        return salary
    def Display(self):
        return f'{super().Display()} {self.CalculateSalary()}'
```
